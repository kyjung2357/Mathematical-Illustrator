```python
# Basic packages
import numpy as np
import matplotlib.pyplot as plt

# font setting
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams["font.family"] = "Times New Roman"
```

### 11.0. 복습

```python
def f(x):
    return x**2

x = np.linspace(-10, 10, 400)

plt.plot(x, f(x))
plt.show()
```

```python
def g(x):
    return np.sqrt(8*x)

x = np.linspace(0, 10, 400)

plt.plot(x, g(x))
plt.plot(x, -g(x))
plt.show()
```

### 11.1. 그래프 꾸미기

```python
def f(x):
    return x**2

x = np.linspace(-10, 10, 400)

plt.figure(figsize=(6, 6))
plt.plot(x, f(x), label='$y = x^2$', color='blue')
plt.xlabel('$x$', fontsize=11)
plt.ylabel('$y$', fontsize=11)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid()
plt.legend()
plt.title('Plot of $y = x^2$', fontsize=11)
plt.show()
```

```python
def f(x):
    return np.sqrt(8*x)

x = np.linspace(0, 10, 400)

plt.figure(figsize=(6, 6))
plt.plot(x, f(x), label='$y^2 = 8x$', color='blue')
plt.plot(x, -f(x), color='blue')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.legend()
plt.title('Plot of $y^2 = 8x$')
plt.show()
```
